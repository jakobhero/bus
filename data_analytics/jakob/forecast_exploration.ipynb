```python
import requests
from datetime import datetime
```

```python
url="https://api.openweathermap.org/data/2.5/onecall"
```

```python
dublin={
    "lat":53.3498,
    "lon":-6.2603
}
```

```python
params={
    "lat":dublin["lat"],
    "lon":dublin["lon"],
    "exclude":"current,minutely,daily",
    "appid":"34eb8ce1e0de10275e7c84882047ff39",
    "units":"metric"
}
```

```python
weather_req=requests.get(url=url,params=params)
weather=weather_req.json()
```

```python
weather
```

```python
hours=[]
for row in weather["hourly"]:
    hours.append(row['dt'])
```

```python
for hour in hours:
    print(datetime.fromtimestamp(hour))
```

```python
(hours[-1]-hours[0])/60/60
```

```python
def get_forecast(timestamp):
    rd_timestamp=timestamp//3600*3600 #round down timestamp to hour
    now=round(datetime.now().timestamp())//3600*3600 #get current hour
    hour_delta=int((timestamp-now)/3600)
    if 0<=hour_delta<48:
        return weather["hourly"][hour_delta]
    else:
        print("Specified time not in 48hrs prediction corridor.")
```

```python
forecast_test=get_forecast(1596478813)
```
